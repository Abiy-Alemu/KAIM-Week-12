Prepare News Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the news page
url = "https://finance.yahoo.com/quote/MSFT/news?p=MSFT"

# Send a GET request to the webpage
response = requests.get(url)

# Parse the page content
soup = BeautifulSoup(response.text, 'html.parser')

# Print the first 500 characters of the HTML to check the structure
print(soup.prettify()[:500])

# Extract news headlines
headlines = []
for item in soup.find_all('h3', class_='Mb(5px)'):  # Modify the class based on the structure
    headline = item.get_text(strip=True)
    headlines.append({'Headline': headline})

# Convert to DataFrame
news_data = pd.DataFrame(headlines)

print(news_data.head())  # Check the extracted headlines
```

```python
if response.status_code == 200:  # Successful request
    data = response.json()
    if 'articles' in data:  # Check if 'articles' exists
        articles = pd.DataFrame(data['articles'])
        news_data = articles[['publishedAt', 'title']]
        news_data.rename(columns={'publishedAt': 'Date', 'title': 'Headline'}, inplace=True)
        news_data['Date'] = pd.to_datetime(news_data['Date']).dt.date
        print(news_data.head())
    else:
        print("No articles found in the response:", data)
else:
    print("API request failed with status code:", response.status_code)
    print("Response:", response.text)

```
